```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.compose import ColumnTransformer

import pandas as pd
import numpy as np

def Numeric_Scaler (df: pd.DataFrame, scaler : StandardScaler,numeric_list_name: list):
    data = df.copy()
    fitted = ColumnTransformer(transformers=[("numerical_scale", scaler, numeric_list_name)], 
                           remainder='passthrough')
    
    ordered_non_numeric = [x for x in df.columns if x not in numeric_list_name]
    
    transformedDf = pd.DataFrame(fitted.fit_transform(data), columns = numeric_list_name+ordered_non_numeric)
    return transformedDf
```

```python
from embedding_encoder import EmbeddingEncoder
from embedding_encoder.utils.compose import ColumnTransformerWithNames

from datawig import Imputer, SimpleImputer
import pickle

Embed_load = EmbeddingEncoder(task = 'classification', 
                    pretrained = True, 
                    mapping_path='../input/tf-embed-feature-categorical-mushroom-data/Embed_TF_Mushromm_Categorical_Data.json')
```

```python


def Feature_Target_Format(df: pd.DataFrame, valY: pd.Series, ifTrain):
    global Embed_load
    newD = df.copy()
    # Relabel the data
    newX = relabled_df(newD)
    
    # Imputer the 'Special Missing' column from pretrained neural network imputer
    # Imputer_Real = Imputer.load("../input/dl-long-mush-stalkroot-na/missing_mushroom")
    # pickle_in = open()
    Imputer_Real = Imputer.load("../input/dl-long-mush-stalkroot-na/missing_mushroom")
    missing_pred = Imputer_Real.predict(newX) # Work on kaggle
#     missing_pred = Imputer_Real.transform(newX)
#     missing_pred = pd.DataFrame(missing_pred['stalk-root'], 
#                                 columns = ['stalk-root_imputed']).set_index(keys= newX.index) # Work on colab

    
    newX = newX.join(missing_pred["stalk-root_imputed"])
    newX["stalk-root"] = np.where(newX["stalk-root"]=='missing', 
                                      newX["stalk-root_imputed"], 
                                      newX["stalk-root"])
    newX.drop(columns='stalk-root_imputed', inplace=True)
#     print(newX.info())
    # print(newX)
    # Format 'two special' column into numeric and boolean 
    newX[['ring-number','bruises']] = newX[['ring-number','bruises']].replace(order_int_list)
#     print(newX.info())
    # Scale the numeric data
    newX = Numeric_Scaler(newX, MinMaxScaler(), ['ring-number'])
#     print(newX.info())
    newX[['ring-number']] = newX[['ring-number']].astype('int64')
    newX[['bruises']] = newX[['bruises']].astype('int64')
#     print(newX.info())
    
    # Format Y_value into boolean
    newY = pd.DataFrame(valY.copy()).replace(order_int_list)['class']
#     newY = valY.copy()
    
    # Embedding Encoder of string Categorical Data using pretrained neural network embedding-encoder
    category_cols = list(newX.columns[(newX.dtypes=='object').values==True])
#     print(newX.info())
#     print(Embed_load)
    filename = 'Embed_Encoder.pkl'
    if ifTrain :
        
        Embed_load.fit(newX[category_cols], newY)
        
        pickle.dump(Embed_load, open(filename, 'wb'))
        print("traiininng")
    else :
        loaded_model = pickle.load(open(filename, 'rb'))
        Embed_load = loaded_model
        
    
    newN = Embed_load.transform(newX[category_cols])
    newN = Numeric_Scaler(newN,scaler = StandardScaler(), numeric_list_name=list(newN.columns))
    newX = pd.concat([newX[['ring-number','bruises']], newN], axis=1)
    

    return newX,newY
```

```python
def CV_Result_Model(list_Model1: dict, verbose : int=4, threshold_ScoreCV: float=0.8, cvFold: int=4):
    model_cv_dta = dict()
    list_Model = list_Model1.copy()
    for name,models in list_Model.items():
        
        
        print("Model {} StartCV".format(name))
        print("+=========================================================================================+")
        cv_dta=cross_val_score(models, Tr_X_tr, TR_Y_tr, cv = cvFold, verbose = verbose)
        model_cv_dta[name] = {"Models DTA":models,"CV_Scores": str(round(cv_dta.mean()*100, 2))+"%" if np.isnan(cv_dta.mean())==False else "~ %", "Score Status": "Good"if cv_dta.mean()>=threshold_ScoreCV else "Bad"}
        print("ModelName: {}, Score_CV: {}".format(name,cv_dta.mean()*100))
        print("+=========================================================================================+\n")
    return pd.DataFrame(model_cv_dta).T
```

#### Checking the baseline Model Performance

```python
from sklearn.metrics import f1_score, recall_score, precision_score
```

```python
def Eval_Baseline_Performance(list_Model: pd.DataFrame, ScoreThreshold: float = 0.8):
    """
    Since we want to predict if the mushroom is edible or poisonous (This crucial if you lost in the wild and except edible plant there are only 
    multiple variety of mushroom that some could give you protein and more food variation for your body to eat and other are poisonous for your body, 
    We want to be able to distinguish between those two by its characteristics), we need our model to have high recall value so that we won't eat the 
    poisonous mushroom to be eaten
    """
    
    copy_list_Model = list_Model.copy()
    
    for x in copy_list_Model["Models DTA"].index:
        copy_list_Model["Models DTA"][x] = copy_list_Model["Models DTA"][x].fit(Tr_X_tr, TR_Y_tr)
    
    concat_list_Model = {
                         "Eval_Precision":dict() ,
                         "Eval_Recall":dict() ,
                         "Eval_F1":dict(), "Precision_Performance":dict(),
                         "Recall_Performance":dict(),
                         "F1_Performance":dict() 
                        }
    for x in copy_list_Model["Models DTA"].index:
        pred_list = copy_list_Model["Models DTA"][x].predict(Tr_X_tst)
        concat_list_Model["Eval_Precision"][x] = precision_score(TR_Y_tst, pred_list, labels=["True Score", "Predict Score"])
        concat_list_Model["Eval_Recall"][x] = recall_score(TR_Y_tst, pred_list, labels=["True Score", "Predict Score"])
        concat_list_Model["Eval_F1"][x] = f1_score(TR_Y_tst, pred_list, labels=["True Score", "Predict Score"])
        concat_list_Model["Precision_Performance"][x] = "Good" if concat_list_Model["Eval_Precision"][x] >= ScoreThreshold else "Bad"
        concat_list_Model["Recall_Performance"][x] = "Good" if concat_list_Model["Eval_Recall"][x] >= ScoreThreshold else "Bad"
        concat_list_Model["F1_Performance"][x] = "Good" if concat_list_Model["Eval_F1"][x] >= ScoreThreshold else "Bad"
    
    newconcat_Data = pd.concat([copy_list_Model, pd.DataFrame(concat_list_Model)], axis=1)
    return newconcat_Data
```
